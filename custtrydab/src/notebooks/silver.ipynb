```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```python
dbutils.widgets.text('catalog_name','feature_dab')
catalog_name = dbutils.widgets.get('catalog_name')
table_name = f'{catalog_name}.slvr.cust_slvr'
```

```python
cust_df = spark.table(f'{catalog_name}.brnz.cust_raw').withColumn('customer_id',col('customer_id').cast('int')).dropDuplicates()
cust_df.write.format('delta').mode('overwrite').option('mergeSchema','true').saveAsTable(table_name)
```

```python
display(spark.sql(f'select * from {table_name}'))
```
