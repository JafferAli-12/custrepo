```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```python
dbutils.widgets.text('catalog_name','feature_dab')
catalog_name = dbutils.widgets.get('catalog_name')
table_name = f'{catalog_name}.gold.customer_summary'
```

```python
cust_df = spark.table(f'{catalog_name}.slvr.cust_slvr')
cust_df.groupBy('name').agg(sum('amount').alias('total_spend')).write.format('delta').mode('overwrite').saveAsTable(table_name)
```

```python
display(spark.sql(f'select * from {table_name}'))
```
